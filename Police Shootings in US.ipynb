# DATA VISUALIZATION OF US POLICE SHOOTING
The main aim of this visualization exercise is to find the determing factors in the shooting that happen in US

```python
import pandas as pd 
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\USER\Downloads\archive.zip")
df.head()
```

```python
df.shape
```

### GENDER

```python
df_gender = df.gender.value_counts()
df_gender
```

```python
colors_list = ['gold', 'black']
explode_list = [0.1, 0] 

df_gender.plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         
                            pctdistance=1.12,   
                            colors=colors_list, 
                         
                            )


plt.title('GENDER RATIO', y=1.12) 

plt.axis('equal') 
# adding  legend
plt.legend(labels=df_gender.index, loc='upper right') 

plt.show()
```

 From the above chart, it can be seen that 95.5% of shootings were males

### RACE

Looking at the recent shooting incidents, it can be seen that race is a major factor in police shooting, from the story of Breonna Taylor to that of George Floyd. It is therefore important to note if this dataset is influenced by race

```python
df.race.value_counts()
df_race = df.race.value_counts()
df_race
```

```python
# plotting data
df_race.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Race') # add to x-label to the plot
plt.ylabel('Number of shootings') # add y-label to the plot
plt.title('SHOOTINGS AND THE RACES AFFECTED') # add title to the plot

plt.show()
```

### AGE

```python
plt.figure(figsize=(7,5))
plt.title('Age Distribution')
sns.distplot(df['age'],kde = False)
```

 It can be seen that the age group majorly affected is from 20 to 40

### TOP 7 CITIES WITH THE MOST SHOOTINGS

```python
topcity = df.city.value_counts().to_frame().reset_index()
topcity.columns = ['City','Shooting Count']
topcity = topcity[0:7] 

plt.title('Top cities with the most shootings')
sns.barplot(data=topcity,x='City',y='Shooting Count')
```

### MENTAL CONDITION OF THE PEOPLE SHOT

```python
df.signs_of_mental_illness.value_counts()
```

```python
df_mentalillness = df.signs_of_mental_illness.value_counts()
colors_list = ['pink', 'green']
explode_list = [0.1, 0] 

df_mentalillness.plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         
                            pctdistance=1.12,   
                            colors=colors_list, 
                         
                            )


plt.title('Chart showing the ratio of people with mental illness', y=1.12) 

plt.axis('equal') 
# adding  legend
plt.legend(labels=df_mentalillness.index, loc='upper right') 

plt.show()
```

77.5% of the people shot were mentally stable

## Suspects trying to flee

```python
df_flee = df.flee.value_counts()
colors_list = ['black', 'purple', 'pink', 'yellow']
explode_list = [0.1, 0] 

df_flee.plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         
                            pctdistance=1.12,   
                            colors=colors_list, 
                         
                            )


plt.title('Chart showing the ratio of people who fleed and who those that did not', y=1.12) 

plt.axis('equal') 
# adding  legend
plt.legend(labels=df_flee.index, loc='upper right') 

plt.show()
```

Most of the people shot were not fleeing

In conclusion, it can be seen that based on this data, race is not a determining factor in police shootings. It can also be noted that its mostly young men that are being affected. So the major determining factors are gender and age.
